```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread('Ro.jpg')
image=cv2.cvtColor(image,cv2.COLOR_RGB2HSV)
background=cv2.resize(cv2.imread('bac.jpg'),(image.shape[1],image.shape[0]))
```

```python
lower_green=np.array([50,50,50])
upper_green=np.array([65,255,255])
mask_green=cv2.inRange(image,lower_green,upper_green)
res=cv2.bitwise_and(image,image,mask=mask_green)

result=image-res
result=cv2.cvtColor(result,cv2.COLOR_HSV2BGR)
result=np.where(result==0 ,background,result)

plt.imshow(result,cmap='gray')
```
